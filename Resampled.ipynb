```python
data = xr.open_dataset('/content/thetao_Omon_CanESM5_historical_r1i1p1f1_gn_199101-200012.nc')
data1 =  xr.open_dataset('/content/thetao_Omon_CanESM5_historical_r1i1p1f1_gn_200101-201012.nc')
data2 =  xr.open_dataset('/content/thetao_Omon_CanESM5_historical_r1i1p1f1_gn_201101-201412.nc')



data.drop_dims('bnds')
data.drop_dims('vertices')

data1.drop_dims('bnds')
data1.drop_dims('vertices')

data2.drop_dims('bnds')
data2.drop_dims('vertices')

data = data.thetao.isel(lev = 0)
data1 = data1.thetao.isel(lev = 0)
data2 = data2.thetao.isel(lev = 0)


ds = xr.concat([data,data1,data2], dim ='time')
ds = ds.rename({'longitude':'lon', 'latitude':'lat'}) 
ds_out = xe.util.grid_global(1, 1)
regridder = xe.Regridder(ds, ds_out, 'bilinear', ignore_degenerate=True, reuse_weights=True, periodic=True)
thetao = regridder(ds)
thetao['time'] = thetao.indexes['time'].to_datetimeindex()
thetao = thetao.drop('lev')
```

```python
lat = thetao.lat.values[:,0]
lon = thetao.lon.values[0,:]

C = xr.Dataset(data_vars= dict(
      thetao = (['time','lat','lon'],thetao.values)), coords = dict(
          lat = (['lat'], lat),
          lon = (['lon'], lon),
          time = thetao.time.values))
C
```

```python
CanESM5_r1i1p1f1_thetao_resampled = C.thetao.interp(lat = Data.latitude.values
                                                    , lon = Data.longitude.values, time = Data.time.values )
CanESM5_r1i1p1f1_thetao_resampled
```
